```
import networkx as nx
from networkx.drawing.nx_pydot import to_pydot
from IPython.core.display import Image

g=nx.DiGraph()
g.add_edge("E","A")
g.add_edge("B","A")
g.add_edge("A","N")
d=to_pydot(g)
d.get_node("N")[0].set_fillcolor("gray")
d.get_node("N")[0].set_style("filled")
d.set_dpi(300)
d.set_margin(0.2)
Image(d.create_png(),width=200)
```
