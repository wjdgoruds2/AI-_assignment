```
import networkx as nx
from IPython.core.display import Image
from networkx.drawing.nx_pydot import to_pydot
g1=nx.Graph()
g1.add_edge("A","B")
g1.add_edge("D","A")
g1.add_edge("B","C")
g1.add_edge("C","D")

d1=to_pydot(g1)
d1.set_dpi(300)
d1.set_margin(0.5)
Image(d1.create_png(),width=300)
```
